```python
import pandas as pd
import os


file_path = r'D:\docs\exercise_data.xlsx'

# Split the path using backslash as the separator (since it's a raw string)
directories = file_path.split('\\')[:-1]  # Exclude the filename

# Print the directories list for debugging
print("Directories:", directories)

# Check if directories list is empty
if directories:
    # Build the directory path incrementally
    current_path = directories[0]  # Start with the drive letter
    for directory in directories[1:]:
        current_path = os.path.join(current_path, directory)
        if not os.path.exists(current_path):
            os.makedirs(current_path)

    try:
        df.to_excel(file_path, index=False)
        print("File saved successfully!")
    except PermissionError:
        print("Error: You don't have permission to write to this location.")
    except Exception as e:
        print(f"An error occurred: {e}")
else:
    print("Error: The file path does not contain any directories.")
```
